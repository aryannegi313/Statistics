```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mean_uofm = 155
sd_uofm = 5
mean_gym = 185
sd_gym = 5
gymperc = 0.30
totalPopSize = 40000

uofm_students = np.random.normal(mean_uofm, sd_uofm, int(totalPopSize * (1 - gymperc)))
students_at_gym = np.random.normal(mean_gym, sd_gym, int(totalPopSize * (gymperc)))

population = np.append(uofm_students, students_at_gym)

plt.figure(figsize=(10,10))

plt.subplot(3,1,1)
sns.distplot(uofm_students)
plt.title("UofM Students only")
plt.xlim([140,200])

plt.subplot(3,1,2)
sns.distplot(students_at_gym)
plt.title("Gym Students only")
plt.xlim([140,200])

plt.subplot(3,1,3)
sns.distplot(population)
plt.title("Full Population")
plt.axvline(np.mean(population))
plt.xlim([140,200])

print()
```

```python
numberSamps = 5000
sampSize = 50

mean_distribution = np.empty(numberSamps)

for i in range(numberSamps):
    randStudents = np.random.choice(population,50)
    mean_distribution[i] = np.mean(randStudents)
    
plt.figure(figsize = (10,8))

plt.subplot(2,1,1)
sns.distplot(population)
plt.title("Full Population")
plt.axvline(np.mean(population))
plt.xlim([140,200])

plt.subplot(2,1,2)
sns.distplot(mean_distribution)
plt.title("Sampling Distribution of the Mean Weight of Gym Goers")
plt.axvline(np.mean(population))
plt.axvline(np.mean(mean_distribution), color='black')
plt.xlim([140,200])
    
```
